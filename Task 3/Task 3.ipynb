```
from transformers import GPT3LMHeadModel, GPT3Tokenizer

# Load the pre-trained GPT-3 model and tokenizer
model_name = 'gpt3'
model = GPT3LMHeadModel.from_pretrained(model_name)
tokenizer = GPT3Tokenizer.from_pretrained(model_name)
```

```
# Function to generate text
def generate_text(prompt, max_length=50):
    input_ids = tokenizer.encode(prompt, return_tensors='pt')
    output = model.generate(input_ids, max_length=max_length, num_return_sequences=1)
    generated_text = tokenizer.decode(output[0], skip_special_tokens=True)
    return generated_text

# Example text generation
prompt = "As a Python development intern, my final task is"
generated_text = generate_text(prompt)
print("Prompt:", prompt)
print("Generated Text:", generated_text)
```

```
# Sample inputs for exploration
prompts = [
    "As a Python development intern, my responsibilities include",
    "In the future, I want to become a skilled Python developer and work on bigger projects"
]

# Generate and analyze text for each prompt
for prompt in prompts:
    generated_text = generate_text(prompt)
    print(f"Prompt: {prompt}")
    print(f"Generated Text: {generated_text}\n")
```

```
# Research Questions

# 1. How well does GPT-3 handle prompts related to a Python development internship?
# 2. What is the creativity level of GPT-3 in generating internship-related text?
# 3. How does GPT-3 perform in generating text for specific professional contexts?
# 4. What are the limitations of GPT-3 in understanding internship-related contexts?
import matplotlib.pyplot as plt

# Data: Number of tokens and corresponding processing times
token_counts = [10, 20, 30, 40, 50]
processing_durations = [0.5, 1.1, 2.2, 3.4, 4.9]

# Plotting the relationship between token count and processing time
plt.figure(figsize=(4, 3.5))
plt.plot(token_counts, processing_durations, linestyle='--', color='teal', marker='s', label='Processing Time')

# Adding labels and title to the plot
plt.xlabel('Number of Tokens', fontsize=12, fontweight='bold')
plt.ylabel('Processing Time (seconds)', fontsize=12, fontweight='bold')
plt.title('Token Count vs. Processing Duration', fontsize=14, fontweight='bold')

# Adding grid and legend for better readability
plt.grid(color='gray', linestyle=':', linewidth=0.5)
plt.legend(loc='upper left', fontsize=10)
plt.tight_layout()

# Display the plot
plt.show()
```
